```python
import pandas as pd

# Memuat file Excel
xls = pd.ExcelFile('dataset_cardio.xlsx')

# Memuat sheet pertama ke dalam DataFrame
nama_sheet = xls.sheet_names[0]  # Pilih nama sheet atau indeks yang diperlukan
df = pd.read_excel(xls, nama_sheet)

# Menanyakan jumlah kelas yang diinginkan
jumlah_kelas = int(input("Mau dibagi berapa kelas? "))

# Membuat list untuk menyimpan data yang dipilih
data_dipilih = []

for i in range(jumlah_kelas):
    # Menanyakan data ke berapa saja yang akan diambil untuk tiap kelas
    data_ambil = input(f"Masukkan indeks data yang akan diambil untuk kelas {i+1} (pisahkan dengan koma): ")
    indeks = [int(x) for x in data_ambil.split(',')]
    data_dipilih.extend(df.iloc[indeks].values.tolist())

# Mengubah list data_dipilih menjadi DataFrame
df_baru = pd.DataFrame(data_dipilih, columns=df.columns)

# Menghapus kolom 'CLASS Cardio'
df_baru = df_baru.drop(columns=['CLASS cardio'])

# Menulis DataFrame baru ke file Excel
with pd.ExcelWriter('output_centroid.xlsx') as writer:
    df_baru.to_excel(writer, sheet_name='Sheet1', index=False)

print("Data berhasil disimpan ke output.xlsx")
```

```python
import pandas as pd
import numpy as np

# Memuat file Excel
xls = pd.ExcelFile('dataset_cardio.xlsx')

# Memuat sheet pertama ke dalam DataFrame
nama_sheet = xls.sheet_names[0]  # Pilih nama sheet atau indeks yang diperlukan
df = pd.read_excel(xls, nama_sheet)

# Menanyakan jumlah centroid yang diinginkan
jumlah_centroid = int(input("Mau dibagi berapa centroid? "))

# Menentukan kolom mana yang harus digunakan (abaikan kolom "data ke")
kolom_fitur = [col for col in df.columns if col.lower() != 'data ke']

# Membuat list untuk menyimpan centroid yang dipilih
centroids = []
centroid_indices = []

for i in range(jumlah_centroid):
    # Menanyakan data ke berapa saja yang akan diambil untuk centroid
    data_ambil = input(f"Masukkan indeks data yang akan diambil untuk centroid {i+1} (pisahkan dengan koma): ")
    indeks = [int(x) for x in data_ambil.split(',')]
    centroid = df.loc[indeks, kolom_fitur].mean()  # Menghitung rata-rata untuk centroid
    centroids.append(centroid)
    centroid_indices.append(indeks)

# Mengubah list centroids menjadi DataFrame
df_centroids = pd.DataFrame(centroids, columns=kolom_fitur)
df_centroids.index = [f'Centroid_{i+1}' for i in range(jumlah_centroid)]

# Memeriksa apakah kolom 'CLASS Cardio' ada di DataFrame
if 'CLASS cardio' in df.columns:
    df = df.drop(columns=['CLASS cardio'])

# Fungsi untuk menghitung jarak Euclidean
def euclidean_distance(row, centroid):
    return np.sqrt(((row - centroid) ** 2).sum())

# Menghitung jarak dari setiap data ke setiap centroid
distances = []
for idx, centroid in df_centroids.iterrows():
    distance = df[kolom_fitur].apply(lambda row: euclidean_distance(row, centroid), axis=1)
    distances.append(distance)

# Menggabungkan jarak ke dalam DataFrame baru
df_distances = pd.DataFrame(distances).T
df_distances.columns = [f'Centroid_{i+1}' for i in range(jumlah_centroid)]

# Menentukan cluster berdasarkan jarak terkecil
df_distances['Cluster'] = df_distances.idxmin(axis=1).apply(lambda x: int(x.split('_')[1]) - 1)

# Menambahkan kolom cluster ke DataFrame asli
df['Cluster'] = df_distances['Cluster']

# Memisahkan data berdasarkan cluster
df_cluster_0 = df[df['Cluster'] == 0]
df_cluster_1 = df[df['Cluster'] == 1]

# Menulis DataFrame baru ke file Excel
with pd.ExcelWriter('output.xlsx') as writer:
    df.to_excel(writer, sheet_name='Data_Awal', index=False)       # Menyimpan data awal
    df_centroids.to_excel(writer, sheet_name='Centroids', index=True)  # Menyimpan centroid
    df_distances.to_excel(writer, sheet_name='Distances', index=False)  # Menyimpan jarak
    df_cluster_0.to_excel(writer, sheet_name='Cluster_0', index=False)  # Menyimpan data cluster 0
    df_cluster_1.to_excel(writer, sheet_name='Cluster_1', index=False)  # Menyimpan data cluster 1

print("Data berhasil disimpan ke output.xlsx")
```

```python
import pandas as pd
import numpy as np

# Memuat file Excel
xls = pd.ExcelFile('dataset_cardio.xlsx')

# Memuat sheet pertama ke dalam DataFrame
nama_sheet = xls.sheet_names[0]  # Pilih nama sheet atau indeks yang diperlukan
df = pd.read_excel(xls, nama_sheet)

# Menanyakan jumlah centroid yang diinginkan
jumlah_centroid = int(input("Mau dibagi berapa centroid? "))

# Menentukan kolom mana yang harus digunakan (abaikan kolom "data ke")
kolom_fitur = [col for col in df.columns if col.lower() != 'data ke']

# Membuat list untuk menyimpan centroid yang dipilih
centroids = []
centroid_indices = []

for i in range(jumlah_centroid):
    # Menanyakan data ke berapa saja yang akan diambil untuk centroid
    data_ambil = input(f"Masukkan indeks data yang akan diambil untuk centroid {i+1} (pisahkan dengan koma): ")
    indeks = [int(x) for x in data_ambil.split(',')]
    centroid = df.loc[indeks, kolom_fitur].mean()  # Menghitung rata-rata untuk centroid
    centroids.append(centroid)
    centroid_indices.append(indeks)

# Mengubah list centroids menjadi DataFrame
df_centroids = pd.DataFrame(centroids, columns=kolom_fitur)
df_centroids.index = [f'Centroid_{i+1}' for i in range(jumlah_centroid)]

# Memeriksa apakah kolom 'CLASS Cardio' ada di DataFrame
if 'CLASS cardio' in df.columns:
    df = df.drop(columns=['CLASS cardio'])

# Fungsi untuk menghitung jarak Euclidean
def euclidean_distance(row, centroid):
    return np.sqrt(((row - centroid) ** 2).sum())

# Menghitung jarak dari setiap data ke setiap centroid
distances = []
for idx, centroid in df_centroids.iterrows():
    distance = df[kolom_fitur].apply(lambda row: euclidean_distance(row, centroid), axis=1)
    distances.append(distance)

# Menggabungkan jarak ke dalam DataFrame baru
df_distances = pd.DataFrame(distances).T
df_distances.columns = [f'Centroid_{i+1}' for i in range(jumlah_centroid)]

# Menentukan cluster berdasarkan jarak terkecil
df_distances['Cluster'] = df_distances.idxmin(axis=1).apply(lambda x: int(x.split('_')[1]) - 1)

# Menambahkan kolom cluster ke DataFrame asli
df['Cluster'] = df_distances['Cluster']

# Memisahkan data berdasarkan cluster
df_cluster_0 = df[df['Cluster'] == 0].copy()
df_cluster_1 = df[df['Cluster'] == 1].copy()

# Fungsi untuk menghitung total nilai dan jumlah data per kolom untuk setiap cluster
def calculate_totals_per_column(df_cluster):
    totals = df_cluster[kolom_fitur].sum()  # Total nilai per kolom
    counts = df_cluster[kolom_fitur].count()  # Jumlah data per kolom
    result = pd.DataFrame({
        'Total_Nilai': totals,
        'Jumlah_Data': counts
    })
    return result

# Menghitung total nilai dan jumlah data untuk setiap cluster
df_totals_cluster_0 = calculate_totals_per_column(df_cluster_0)
df_totals_cluster_1 = calculate_totals_per_column(df_cluster_1)

# Menghitung rasio
df_totals_cluster_0['Rasio'] = df_totals_cluster_0['Total_Nilai'] / df_totals_cluster_0['Jumlah_Data']
df_totals_cluster_1['Rasio'] = df_totals_cluster_1['Total_Nilai'] / df_totals_cluster_1['Jumlah_Data']

# Menggunakan rasio sebagai centroid baru
new_centroids = [df_totals_cluster_0['Rasio'], df_totals_cluster_1['Rasio']]

df_new_centroids = pd.DataFrame(new_centroids, columns=kolom_fitur)
df_new_centroids.index = ['New_Centroid_0', 'New_Centroid_1']

# Menghitung jarak dari setiap data ke centroid baru
distances_new = []
for idx, centroid in df_new_centroids.iterrows():
    distance = df[kolom_fitur].apply(lambda row: euclidean_distance(row, centroid), axis=1)
    distances_new.append(distance)

# Menggabungkan jarak ke dalam DataFrame baru
df_distances_new = pd.DataFrame(distances_new).T
df_distances_new.columns = ['New_Centroid_0', 'New_Centroid_1']

# Menentukan cluster baru berdasarkan jarak terkecil
df_distances_new['New_Cluster'] = df_distances_new.idxmin(axis=1).apply(lambda x: int(x.split('_')[2]))

# Menambahkan kolom cluster baru ke DataFrame asli
df['New_Cluster'] = df_distances_new['New_Cluster']

# Menyimpan data ke file Excel
with pd.ExcelWriter('output.xlsx') as writer:
    df.to_excel(writer, sheet_name='Data_Awal', index=False)       # Menyimpan data awal
    df_centroids.to_excel(writer, sheet_name='Centroids', index=True)  # Menyimpan centroid
    df_distances.to_excel(writer, sheet_name='Distances', index=False)  # Menyimpan jarak
    df_cluster_0.to_excel(writer, sheet_name='Cluster_0', index=False)  # Menyimpan data cluster 0
    df_cluster_1.to_excel(writer, sheet_name='Cluster_1', index=False)  # Menyimpan data cluster 1
    df_totals_cluster_0.to_excel(writer, sheet_name='Totals_Cluster_0')  # Menyimpan total dan rasio cluster 0
    df_totals_cluster_1.to_excel(writer, sheet_name='Totals_Cluster_1')  # Menyimpan total dan rasio cluster 1
    df_new_centroids.to_excel(writer, sheet_name='New_Centroids', index=True)  # Menyimpan centroid baru
    df_distances_new.to_excel(writer, sheet_name='New_Distances', index=False)  # Menyimpan jarak ke centroid baru
    df.to_excel(writer, sheet_name='Data_With_New_Cluster', index=False)  # Menyimpan data dengan cluster baru

print("Data berhasil disimpan ke output.xlsx")
```
